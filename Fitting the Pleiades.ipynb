# Fitting the Pleiades

In this workbook I will show how to retrieve data from the Hipparcos catalog and Vizier, to plot them in a Hertzsprung-Russell Diagram (or a color magnitude diagram -- for observers), and how to overplot isochrones to fix the location of the main sequence and to determine the age and distance to the Pleiades.

```python
%pylab inline
import numpy as np
import pandas as pd
```

```python
from IPython.display import Image
Image(url='http://astropixels.com/openclusters/images/M45-01w.jpg')
```

To retrieve data from the Hipparcos catalogue, go to http://vizier.cfa.harvard.edu/viz-bin/VizieR?-source=I/239/hip_main.  This will give you a window where you can select which data you want to download. For this excercise I selected the following columns: HIP, Vmag, Plx (parallax), B-V and SPtype. On the top of the page, for Target fill in M45 for object name and 2 degrees for search radius. Under "Preferences" (left side of page) you pick max: "unlimited" and in the dropdown immediately below that pick "; - separated-values". Submit and you will download a file, should be about 55 entries. 

If you want to make sure that all the stars in this file really belong to the Pleiades, make sure to also check "pmRA" and "pmDE". Since all the stars in the Pleiades are born at the same time, out of the same gas cloud, you would expect them to have a similar motion though space. This is expressed by the "proper motion," a measure of the relative speed of the stars with respect to the earth. As you can see, the proper motion is generally in the range of 17-23 milli arcseconds/year for Right Ascension and 41-46 milli arcseconds/year for Declination. Anything that is (far) outside of these ranges can be discarded. In this way you can also increase the search radius (in the example above it was set for 2 degrees) to a value that is much larger, because you can easily filter out stars that do not have a proper motion that is in the right range.

```python
filename = 'M45.tsv'
df = pd.read_table(filename, skiprows=48, sep=';', header=None, index_col=0,
                   names = ['HIP', 'Vmag', 'Plx', 'B-V', 'SpType'],
                   skip_footer=1)
```

I read the file in with the Python library "Pandas" as it is really good in dealing with tables.

Just to check the data I display the first 5 lines with the head command, and the last 5 lines with the tail command.

```python
df.head()
```

```python
df.tail()
```

To clean up the data I fill all the blank spots in the talbe with a NaN (Not a Number) and then I drop these lines with the command .dropna( ). I inspect the data with the describe command.

```python
df_clean = df.applymap(lambda x: np.nan if isinstance(x, basestring)
                       and x.isspace() else x)
```

```python
df_clean= df_clean.dropna()
df_clean.describe()
```

To make sure that I can do calculations with the numbers in the table, I assign the "float" type to the magnitude and distance column. Then I calculate the Absolute Magnitude ($M_V$) from the apparent magnitude ($V$) and the distance (parallax in parsec).

```python
df_clean['Vmag'] = df_clean['Vmag'].astype(np.float)
df_clean['Plx'] = df_clean['Plx'].astype(np.float)
df_clean['B-V'] = df_clean['B-V'].astype(np.float)
```

```python
# Add a new column with the absolute magnitude
df_clean['M_V'] = df_clean['Vmag'] + 5 * log10(df_clean['Plx']/100.)
```

Check whether whether we indeed have a new column.

```python
df_clean.head()
```

Allright, let's plot our data. The commands below may sound complex, but I first define a figure with a specific size, and then I add a plot to that figure and call that "ax". From then on I manipulate ax with titles, labels and axes. 

```python
fig = figure(figsize=(8,10))
ax = fig.add_subplot(111)

ax.set_xlim(-0.5, 2.5)
ax.set_ylim(15, -10)
ax.grid()
ax.set_title('H-R Diagram Pleiades \n (Hipparcos catalog)')

ax.title.set_fontsize(20)
ax.set_xlabel('Color index B-V')
ax.xaxis.label.set_fontsize(20)
ax.set_ylabel('Absolute magnitude')
ax.yaxis.label.set_fontsize(20)

ax.scatter(df_clean['B-V'], df_clean['M_V'],
#           s=50, edgecolors='none', alpha=0.015, c='k')
           s=10, edgecolors='none', c='k')

ax.tick_params(axis='both', labelsize=14)
```

That looks really nice! We can see a clear Main Sequence (MS) and two stars that are probably on the Giant Branch. The upper end of the Main Sequence is also well defined, and shows a hint of stars curving away from the MS.

To determine the age of the cluster we can overplot some isochrones. For more info about isochrones, check out this website: http://angryastronomer.blogspot.com/2007/04/stellar-evolution-isochrones.html (unfortunately the figures are missing). To download theoretical isochrones, you can go to many different sources. Some well known databases are the <a href="http://stellar.dartmouth.edu/models/">Dartmouth Stellar Evolution Database</a>, <a href="http://stev.oapd.inaf.it/cgi-bin/cmd">the Padova stellar evolutionary tracks and isochrones</a>, the <a href="http://obswww.unige.ch/Recherche/evol/Geneva-grids-of-stellar-evolution">Geneva Grids of Stellar Evolution</a>, the <a href="http://basti.oa-teramo.inaf.it/index.html">BASTI: A Bag of Stellar Tracks and Isochrones...</a> and <a href="http://www.astro.ulb.ac.be/~siess/pmwiki/pmwiki.php/WWWTools/Isochrones">Lionell Siess' isochrones</a>. I choose to download them from the following link:
http://pleiadi.pd.astro.it/isoc_photsys.00/isoc_ubvrijhk/index.html

Reading the data in is easy :) I choose a set that has a metallicity that is similar to the sun (z=0.019).

```python
loga, logl, logte = np.genfromtxt('isoc_z019.dat',
  usecols=(0, 3, 4), unpack=True)
```

If you download the file for yourself, you can see that the file contains several big tables, organized by age and mass. To pick out the models that have a certain age, I do a simple numpy search that returns the index values for the lines that correspond to this criterium. These indeces are now stored in w7, w8, w9 and w10.

```python
w7 = numpy.where(loga == 7.0)
w8 = numpy.where(loga == 8.0)
w9 = numpy.where(loga == 9.0)
w10 = numpy.where(loga == 10.0)
```

Let's plot them and see what we get!

```python
plot(logte[w7], logl[w7],label='10 Myr')
plot(logte[w8], logl[w8],label='100 Myr')
plot(logte[w9], logl[w9],label='1000 Myr')
plot(logte[w10], logl[w10],label='10000 Myr')
plt.xlabel('Log(Teff)')
plt.ylabel('Log L')
plt.axis([4.6, 3.4, 0, 6])
plt.legend(loc='lower left')
```

That looks great and exactly what we would expect. The younger the cluster is, the further on the MS it goes. Older isochrones "peel" off the main sequence and move toward the red part of the Hertzsprung Russell Diagram. While these look great, we can't really use them to overplot our "observations" since these are normally expressed as magnitudes and not as luminosities and temperatures. Fortunately, the isochrone file also has Mb and Mv, so we can also calculate B-V.

```python
loga, M_B, M_V = np.genfromtxt('isoc_z019.dat',
  usecols=(0, 8, 9), unpack=True)
BminV=M_B-M_V
```

Let's define the ages for our isochrones again.

```python
J7 = numpy.where(loga == 7.0)
J8 = numpy.where(loga == 8.00)
J9 = numpy.where(loga == 9.0)
J10 = numpy.where(loga == 10.0)

```

```python
plot(BminV[J7], M_V[J7],label='10 Myr')
plot(BminV[J8], M_V[J8],label='100 Myr')
plot(BminV[J9], M_V[J9],label='1000 Myr')
plot(BminV[J10], M_V[J10],label='10000 Myr')
plt.xlabel('B-V')
plt.ylabel('V')
plt.axis([-0.5, 2.0, 8, -8])
plt.legend(loc='lower left')
```

Since we plot the isochrones now on a color-magnitude diagram (CMD) they look a little bit different, but the main shape is still there. Let's overplot the data for the Pleiades.

```python
fig = figure(figsize=(8,10))
ax = fig.add_subplot(111)

ax.set_xlim(-0.5, 2.5)
ax.set_ylim(15, -10)
ax.grid()
ax.set_title('H-R Diagram Pleiades \n (Hipparcos catalog)')

ax.title.set_fontsize(20)
ax.set_xlabel('Color index B-V')
ax.xaxis.label.set_fontsize(20)
ax.set_ylabel('Absolute magnitude')
ax.yaxis.label.set_fontsize(20)

ax.scatter(df_clean['B-V'], df_clean['M_V'],
#           s=50, edgecolors='none', alpha=0.015, c='k')
           s=10, edgecolors='none', c='k')
ax.plot(BminV[J7], M_V[J7],label='10 Myr')
ax.plot(BminV[J8], M_V[J8],label='100 Myr')
ax.plot(BminV[J9], M_V[J9],label='1000 Myr')
ax.plot(BminV[J10], M_V[J10],label='10000 Myr')
ax.tick_params(axis='both', labelsize=14)
ax.legend(loc='lower left')
```

From this plot you can see that the age of the cluster is approximately 100 Myr, as that is the isochrone that fits the best. It is probably a little bit older, because that will shift the isochrone a little bit to the right. The most accurate techniques to determine the age of the cluster give 115 Myr!

Since the data for the Pleiades is pretty sparse (the total cluster is supposed to have ~400 stars), I decided to see whether I could find better data. The best way to find BV photometry data is through  <a href="http://vizier.cfa.harvard.edu/viz-bin/VizieR">ViZier</a>. Type "BV Photometry" in the first box and optionally an object in the lower box and then click find catalogs on the lower left, and you should get a (long) list or possible resources. Pick one that you like (the number of stars for which there is data is given by the small numbers just before the title of each entry).

```python
filename = 'm45.dat'
df2 = pd.read_table(filename, skiprows=2, header=None, index_col=0,
                   names = ['No', 'Ref', 'Vmag', 'B-V', 'U-B','N'])
```

Reading in the data. We don't have to worry about U-B (it has less data than the rest --probably white spaces, so it could potentially give problems).

```python
df2.shape
df2.describe()
```

This time we don't have any information about the distance of the cluster. So, let's plot it anyway and see what happens.

```python
fig = figure(figsize=(8,10))
ax = fig.add_subplot(111)

ax.set_xlim(-0.5, 2.5)
ax.set_ylim(15, -10)
ax.grid()
ax.set_title('H-R Diagram Pleiades')

ax.title.set_fontsize(20)
ax.set_xlabel('Color index B-V')
ax.xaxis.label.set_fontsize(20)
ax.set_ylabel('Absolute magnitude')
ax.yaxis.label.set_fontsize(20)

ax.scatter(df2['B-V'], df2['Vmag'],
#           s=50, edgecolors='none', alpha=0.015, c='k')
           s=10, edgecolors='none', c='k')
ax.plot(BminV[J7], M_V[J7],label='10 Myr')
ax.plot(BminV[J8], M_V[J8],label='100 Myr')
ax.plot(BminV[J9], M_V[J9],label='1000 Myr')
ax.plot(BminV[J10], M_V[J10],label='10000 Myr')
ax.tick_params(axis='both', labelsize=14)
ax.legend(loc='lower left')
```

What you see is that the MS ends up a lot lower than what we would expect from the isochrones. The reason for that is that what we plot on the y-axis is not absolute magnitude ($M_v$), but apparent magnitude $V$ or $m_v$. They are related, however, through the distance modulus, given by $$m_v - M_v = -5 + 5\log_{10} d$$ with $d$ given in parsec. The difference between the absolute and the apparent magnitude can thus be rewritten to give the distance $$d=10^{(m_v-M_v+5)/5}.$$

Using the graph, we can estimate the distance modulus $m_v - M_v$ to be 5.67, which gives a distance to the cluster of 136 pc. A plot with the right distance taken into account is shown below.

```python
fig = figure(figsize=(8,10))
ax = fig.add_subplot(111)

ax.set_xlim(-0.5, 2.5)
ax.set_ylim(15, -10)
ax.grid()
ax.set_title('H-R Diagram Pleiades')

ax.title.set_fontsize(20)
ax.set_xlabel('Color index B-V')
ax.xaxis.label.set_fontsize(20)
ax.set_ylabel('Absolute magnitude')
ax.yaxis.label.set_fontsize(20)

ax.scatter(df2['B-V'], df2['Vmag']-5.67,
#           s=50, edgecolors='none', alpha=0.015, c='k')
           s=10, edgecolors='none', c='k')
ax.plot(BminV[J7], M_V[J7],label='10 Myr')
ax.plot(BminV[J8], M_V[J8],label='100 Myr')
ax.plot(BminV[J9], M_V[J9],label='1000 Myr')
ax.plot(BminV[J10], M_V[J10],label='10000 Myr')
ax.tick_params(axis='both', labelsize=14)
ax.legend(loc='lower left')
```

As you can see, this is a very different shape of stellar population on the color magnitude diagram. The main reason for this is that M3 is much older than the Pleides and thus most of the Main Sequence is gone. Many more stars are now on the Giant Branch, the Horizontal Branch and the Asymptotic Giant Branch: all of them are realy well defined!
